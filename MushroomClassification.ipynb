```python
from sklearn.model_selection import train_test_split

X = data[data.columns.difference(['class'])]
y = data['class']

# Basic split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify = y)

print(f"Basic split shapes:")
print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")
```

```python
#!pip install ydf
```

```python
train_ds = X_train.assign(class_label=y_train.values)
test_ds = X_test.assign(class_label=y_test.values)
```

```python
import ydf

model = ydf.RandomForestLearner(label="class_label",
                                task=ydf.Task.CLASSIFICATION).train(train_ds)
```

```python
evaluation = model.evaluate(test_ds)
evaluation
```
